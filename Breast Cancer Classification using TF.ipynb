```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.33)
n, d=x_train.shape
print(n,d)
```

```
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```
import tensorflow as tf
model=tf.keras.models.Sequential() 
model.add(tf.keras.layers.Dense(1,input_shape=(d,),activation='sigmoid'))
```

```
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
y=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100) 
```

```
 import matplotlib.pyplot as plt
plt.plot(y.history['accuracy'], label='acc')
plt.plot(y.history['val_accuracy'], label='val_acc')
plt.legend();
```

```
plt.plot(y.history['loss'], label='loss')
plt.plot(y.history['val_loss'], label='val_loss')
plt.legend();
```

```
y=model.evaluate(x_train,y_train)
o=model.evaluate(x_test, y_test)
y1=float(y[1])*100
o1=float(o[1])*100
print('Train Score:',y1,'%')
print('Test Score:',o1,'%' )
```
